```python
def check_data_exists(till_date='2031-04-01'):
    from google.cloud import bigquery
    client = bigquery.Client('ltvsubscribers')
    till_date='2031-04-01'
    sql="""
    select * from `ltvsubscribers.anit_sandbox.pt_ltv_subs_trial_period_by_quarter` 
    where day_dt= @till_date
    """
    job_config = bigquery.QueryJobConfig(
            query_parameters=[
                bigquery.ScalarQueryParameter("till_date", "TIMESTAMP", till_date),
            ]
        )
    df = client.query(sql, job_config=job_config).to_dataframe()
    
    if len(df)==0:
        return False
    else:
        return True
```
